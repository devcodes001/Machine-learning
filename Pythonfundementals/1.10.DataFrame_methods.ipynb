#Importand DataFrame methods

```python
import numpy as np
import pandas as pd
marks = pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]
],columns=['iq','marks','package'])

marks
```

```python
#value_counts
marks['marks'].value_counts()
```

```python
#sort_values
marks['iq'].sort_values()
```

```python
marks['iq'].sort_values(ascending = False)
```

```python
#rank()
marks['rank'] = marks['marks'].rank(ascending = False)
marks
```

```python
#sort_index()
marks.sort_index(ascending=False)
```

```python
#set_index()
marks.set_index('iq')
```

```python
#rename()
marks.rename(columns = {'marks' : 'mark'})
```

```python
#reset_index()
marks.reset_index(drop =True)
```

```python
# unique & nunique
marks['marks'].unique()
marks['marks'].nunique()
```

```python
# isnull/notnull/hasnans
marks.isnull().sum()
```

```python
marks.notnull().sum()
```

```python
marks.dropna()
```

```python
marks.fillna(marks.mean())
```

```python
# drop_duplicates
marks.drop_duplicates()
```

```python
#drop
marks.drop(index = [1,2])
marks.drop(columns = ['marks','rank'])
```

```python
#apply
marks['package'].apply(lambda x: x*10)
```

```python
marks.isin([100,70])
```

```python
marks.corr()
```

```python
#DataFrameName.insert(loc, column, value, allow_duplicates = False)
marks.insert(0,'age',[12,13,14,15,11])
```
